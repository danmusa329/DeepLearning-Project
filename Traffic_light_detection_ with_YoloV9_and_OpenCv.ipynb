```python
#inference with opencv on a video
import cv2
import numpy as np
from ultralytics import YOLO
# Load YOLOv9 model
#model = ("runs/detect/train/weights/best.pt")
import torch

# Load the YOLOv9 model
model_1 = YOLO("runs/detect/train/weights/best.pt")

# Open video file
video_path = 'movie1.mp4'
cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print("Error: Could not open video.")
    exit()

# Get video properties
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))

# Define the codec and create VideoWriter object
output_path = 'annotated_movie2.mp4'
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # You can also use 'XVID' or other codecs
out = cv2.VideoWriter(output_path, fourcc, fps, (frame_width, frame_height))

# Loop through video frames
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    
    # Prepare frame for YOLOv9
    img = frame[..., ::-1]  # BGR to RGB
    results = model_1(img)

    # Print results to understand its structure
    #print(results)

    # Check if results is a list and process detections
    if isinstance(results, list) and len(results) > 0:
        boxes = results[0].boxes  # Access the Boxes object from the first element
        if boxes is not None:
            for box in boxes:
                x1, y1, x2, y2 = map(int, box.xyxy[0].tolist())  # Bounding box coordinates
                conf = box.conf[0].item()  # Confidence score
                cls = int(box.cls[0].item())  # Class ID
                label = f"{results[0].names[cls]} {conf:.2f}"
                
                # Draw bounding box and label on the frame
                cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
                cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    # Write the frame into the file
    out.write(frame)

# Release resources
cap.release()
out.release()

print(f"Annotated video saved as {output_path}")

```

```python
#inference with opencv using webcam
import cv2
import numpy as np
from ultralytics import YOLO

# Load the YOLOv9 model
model_1 = YOLO("runs/detect/train/weights/best.pt")

# Open webcam
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("Error: Could not open webcam.")
    exit()

# Get webcam properties
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))

# Loop through webcam frames
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    
    # Prepare frame for YOLOv9
    img = frame[..., ::-1]  # BGR to RGB
    results = model_1(img)

    # Check if results is a list and process detections
    if isinstance(results, list) and len(results) > 0:
        boxes = results[0].boxes  # Access the Boxes object from the first element
        if boxes is not None:
            for box in boxes:
                x1, y1, x2, y2 = map(int, box.xyxy[0].tolist())  # Bounding box coordinates
                conf = box.conf[0].item()  # Confidence score
                cls = int(box.cls[0].item())  # Class ID
                label = f"{results[0].names[cls]} {conf:.2f}"
                
                # Draw bounding box and label on the frame
                cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
                cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    # Display the frame
    cv2.imshow('Webcam', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release resources
cap.release()
cv2.destroyAllWindows()
```
